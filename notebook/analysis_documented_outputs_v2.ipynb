# About

This notebook contains the main analytical workflow for a university research project on the **economic feasibility of Vehicle-to-Grid (V2G) systems in Hungary**.

## Project motivation
Hungary is experiencing rapid growth in renewable electricity production, especially from solar PV. This creates structural imbalances in the electricity system, including low midday prices, higher evening prices, and rising balancing needs. V2G may help address these patterns by allowing EV batteries to act as distributed storage.

## Research objectives
The notebook develops a data-driven microeconomic assessment of V2G in Hungary. The main objectives are:
- estimate arbitrage potential from charging at low-price hours and discharging at high-price hours,
- quantify grid-side annual income, cost, and profit,
- evaluate projected EV fleet growth and projected infrastructure costs,
- forecast electricity prices through 2030,
- assess 10-year financial feasibility, including incentives and NPV.

## Workflow
The analysis is organized into the following stages:
1. data ingestion from electricity-market, EV, battery-capacity, and infrastructure-cost datasets,
2. cleaning and harmonization of modeling inputs,
3. exploratory analysis of electricity-price dynamics,
4. forecasting of EV stock, electricity prices, and infrastructure costs,
5. simplified V2G economic modeling.

## Output management
The notebook now uses an output folder:

`../output/`

Selected charts are saved there as **PNG** files and selected reporting tables are saved there as **CSV** files in addition to being shown in the notebook.

```python
# Create normalized copies for matching
def normalize_model_name(text):
    text = str(text).lower().strip()
    text = re.sub(r'\s+', ' ', text)
    return text

# Normalized copies
ev_type_tmp = ev_type_df.copy()
battery_tmp = battery_capacity_df.copy()

ev_type_tmp['model_key'] = ev_type_tmp['EV Model'].apply(normalize_model_name)
battery_tmp['model_key'] = battery_tmp['EV Model'].apply(normalize_model_name)

# Manual standardization
ev_type_tmp['model_key'] = ev_type_tmp['model_key'].replace({
    'byd sealion7': 'byd sealion 7'
})

battery_tmp['model_key'] = battery_tmp['model_key'].replace({
    'byd sealion7': 'byd sealion 7'
})

def find_matches(ev_key, battery_series):
    # Match as a phrase, not as a raw substring
    pattern = rf'(^|[^a-z0-9]){re.escape(ev_key)}([^a-z0-9]|$)'
    return battery_series.str.contains(pattern, regex=True, na=False)
    
# find all battery dataset rows where the battery model contains the top model
matches = []
for _, ev_row in ev_type_tmp.iterrows():
    ev_model = ev_row['EV Model']
    units_sold = ev_row['Unit Sold 2025']
    ev_key = ev_row['model_key']
    
    mask = find_matches(ev_key, battery_tmp['model_key'])
    matched_rows = battery_tmp[mask].copy()
    
    if matched_rows.empty:
        matches.append({
            'EV Model': ev_model,
            'Unit Sold 2025': units_sold,
            'Matched Battery Model': None,
            'Capacity': None
        })
    else:
        for _, bat_row in matched_rows.iterrows():
            matches.append({
                'EV Model': ev_model,
                'Unit Sold 2025': units_sold,
                'Matched Battery Model': bat_row['EV Model'],
                'Capacity': bat_row['Capacity']
            })

ev_type_matches_df = pd.DataFrame(matches)
ev_type_matches_df
```

```python
# summarize to one capacity per top-selling model

ev_profiles_df = (
    ev_type_matches_df
    .groupby(['EV Model', 'Unit Sold 2025'], as_index=False)
    .agg(
        Matched_Models=('Matched Battery Model', lambda x: ', '.join(x.dropna().astype(str))),
        Battery_Capacity=('Capacity', 'mean')
    )
)

# manualy add battery_capacity for 'BYD YUAN PLUS / ATTO-3' = average (49.92 kWh and 60.48 kWh)
ev_profiles_df.loc[
    ev_profiles_df['EV Model'] == 'BYD YUAN PLUS / ATTO-3',
    'Battery_Capacity'
] = (49.92 + 60.48) / 2

display(ev_profiles_df)
save_table(ev_profiles_df, 'report_ev_profiles_top10', index=False)
```

Drill down to specific dates:

```python
def plot_price_window_with_daily_average(price_df, start_date, end_date, figsize=(16, 6), save_name=None):
    """
    Plot hourly electricity prices between start_date and end_date,
    with a daily average line shown separately for each day.

    Parameters
    ----------
    price_df : pd.DataFrame
        Must contain:
        - 'Datetime' as datetime64
        - 'Price' as numeric
    start_date : str or datetime-like
        Example: '2021-01-01'
    end_date : str or datetime-like
        Example: '2021-01-07'
    figsize : tuple
        Matplotlib figure size
    """
    df = price_df.copy()
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    end_datetime = end_date + pd.Timedelta(days=1) - pd.Timedelta(hours=1)

    df = df[
        (df['Datetime'] >= start_date) &
        (df['Datetime'] <= end_datetime)
    ].copy()

    if df.empty:
        print('No data available in the selected date range.')
        return

    df = df.sort_values('Datetime').copy()
    df['date'] = df['Datetime'].dt.date
    df['daily_avg_price'] = df.groupby('date')['Price'].transform('mean')

    fig, ax = plt.subplots(figsize=figsize)

    ax.plot(df['Datetime'], df['Price'], marker='o', linewidth=1.5, label='Hourly Price')

    for _, day_df in df.groupby('date'):
        ax.plot(
            day_df['Datetime'],
            day_df['daily_avg_price'],
            linestyle='--',
            linewidth=2,
            color='red'
        )

    ax.fill_between(
        df['Datetime'],
        df['Price'],
        df['daily_avg_price'],
        where=(df['Price'] < df['daily_avg_price']),
        alpha=0.2,
        color='green',
        label='Charging Opportunity'
    )

    ax.fill_between(
        df['Datetime'],
        df['Price'],
        df['daily_avg_price'],
        where=(df['Price'] > df['daily_avg_price']),
        alpha=0.3,
        color='red',
        label='Discharging Opportunity'
    )

    ax.plot([], [], linestyle='--', linewidth=2, color='red', label='Daily Average')

    ax.set_title(f'Electricity Price Profile from {start_date.date()} to {end_date.date()}')
    ax.set_xlabel('Datetime')
    ax.set_ylabel('Price (EUR/kWh)')
    ax.grid(True)

    ax.xaxis.set_major_locator(mdates.DayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d\n%H:%M'))

    plt.xticks(rotation=45)
    ax.legend()
    plt.tight_layout()
    if save_name is not None:
        save_plot(save_name)
    plt.show()
```

# Modeling

We start by partitioning data for year 2021, which we have essential data available.

Later we will extend our model to 2030, by timeseries prediction.

## Assumptions

- Consumer price is equal to generation, transmission and distribution value of Electricity. This is very big statement, but let's assume it.
- The top 10 units sold could be used to calculate the average battery capacity.
- BEV and PHEV are all EV.
- No technical barriers for EV about the duration and pace of discharge and charge phases. However we assume linear discharge and charge.
- Recharge after midnight fully restores the battery before morning departure.
- Conversion from common charging station into V2G station is the only cost of this project.
- Consider 'mid2021' in the station_conversion_cost_df in this model.


## Parameters:

- ev_param_v2g_compatible{v2g_station, ev_compatible, preferance}: The percentage of EV owners available V2G charging station at home, EV compatible with V2G, and prefer to participate in V2G
- ev_param_v2g_available{workday, weekend}: The percentage of EV owners pluged-in to station in workday and weekend during early evening hours
- ev_param_plugin_time: Based on EV owner behaviour, the time in early evening the EV could be pluged-in to station.
- ev_param_plugin_battery_percentage: The percentage of battery at the moment of pluging-in.
- ev_param_plugout_time: Based on EV owner behaviour, the time in the morning the EV could be pluged-out from station.
- ev_param_safe_capacity: The minimum capacity of battery to not discharge into grid.
- spread_treshold: #todo
- amortization_years: Amortization of investment. Default must assume 0.


## Objective

- The difference between discharge in early evening and charge during midnight must be calculated. It is interpreted as income for the grid.
- The profit must be calculated by income - cost

```python
ev_param_v2g_compatible = {
    'v2g_station': 0.9, 
    'ev_compatible': 0.9, 
    'preferance': 0.9 
}

ev_param_v2g_available = {
    'workday': 0.9, 
    'weekend': 0.5
}

ev_param_plugin_time = 18      # early evening
ev_param_plugout_time = 7      # morning
ev_param_plugin_battery_percentage = 0.6667
ev_param_safe_capacity = 0.3334
amortization_years = 10
```

## Calculations

### Prepare EV stock and battery inputs for 2021

```python
# EV stock in 2021
ev_2021 = ev_count_df.loc[ev_count_df['Year'] == 2021, 'EV'].iloc[0]

# Average battery capacity from top 10 best-selling EVs
weighted_avg_battery_capacity_kwh = (ev_profiles_df['Unit Sold 2025'] * ev_profiles_df['Battery_Capacity']).sum() / ev_profiles_df['Unit Sold 2025'].sum()

print('EV stock in 2021:', ev_2021)
print('Weighted average battery capacity (kWh):', round(weighted_avg_battery_capacity_kwh, 2))
```

### Prepare station conversion cost

```python
# Separate capex and annual opex
annual_metering_cost = station_conversion_cost_df.loc[
    station_conversion_cost_df['cost'] == 'Metering equipment',
    'mid2021'
].iloc[0]

capex_per_station = station_conversion_cost_df.loc[
    station_conversion_cost_df['cost'] != 'Metering equipment',
    'mid2021'
].sum()

annualized_capex_per_station = capex_per_station / amortization_years
annual_cost_per_station = annualized_capex_per_station + annual_metering_cost

print('CAPEX per station (EUR):', capex_per_station)
print('Annual metering cost per station (EUR/year):', annual_metering_cost)
print('Annualized CAPEX per station (EUR/year):', round(annualized_capex_per_station, 2))
print('Annual total cost per station (EUR/year):', round(annual_cost_per_station, 2))
```

### Model-compatible EV pool

```python
v2g_compatibility_rate = (
    ev_param_v2g_compatible['v2g_station']
    * ev_param_v2g_compatible['ev_compatible']
    * ev_param_v2g_compatible['preferance']
)

eligible_ev_2021 = ev_2021 * v2g_compatibility_rate

print('Compatibility rate:', round(v2g_compatibility_rate, 4))
print('Eligible EVs in 2021:', round(eligible_ev_2021))
```

### Build daily price features for 2021

```python
price_2021_df = price_df[price_df['Datetime'].dt.year == 2021].copy()

price_2021_df['date'] = price_2021_df['Datetime'].dt.date
price_2021_df['hour'] = price_2021_df['Datetime'].dt.hour
price_2021_df['weekday'] = price_2021_df['Datetime'].dt.weekday
price_2021_df['day_type'] = np.where(price_2021_df['weekday'] < 5, 'workday', 'weekend')
```

### Extract daily charge and discharge prices

```python
daily_rows = []

for current_date, day_df in price_2021_df.groupby('date'):
    day_df = day_df.sort_values('hour').copy()

    # Daily average price over all 24 hours
    daily_avg_price = day_df['Price'].mean()

    # Discharge window: 18:00 to 00:00
    # We consider hours 18, 19, 20, 21, 22, 23
    # Keep only hours with price above that day's average
    evening_df = day_df[
        (day_df['hour'] >= ev_param_plugin_time) &
        (day_df['hour'] < 24)
    ].copy()

    discharge_hours_df = evening_df[
        evening_df['Price'] > daily_avg_price
    ].copy()

    if discharge_hours_df.empty:
        avg_discharge_price = np.nan
        discharge_hours_count = 0
    else:
        avg_discharge_price = discharge_hours_df['Price'].mean()
        discharge_hours_count = len(discharge_hours_df)

    # Charge window: 22:00 to 07:00 next morning
    # We approximate this as:
    # same date hours 22, 23
    # plus next date hours 0..6
    # -----------------------------
    next_date = current_date + pd.Timedelta(days=1)

    # Next day daily average price over all 24 hours
    next_day_avg_price = price_2021_df[
        price_2021_df['date'] == next_date
    ]['Price'].mean()

    same_night_df = price_2021_df[
        (price_2021_df['date'] == current_date) &
        (price_2021_df['hour'] >= 22)
    ][['Datetime', 'date', 'hour', 'Price']].copy()

    next_morning_df = price_2021_df[
        (price_2021_df['date'] == next_date) &
        (price_2021_df['hour'] < ev_param_plugout_time)
    ][['Datetime', 'date', 'hour', 'Price']].copy()

    charge_window_df = pd.concat([same_night_df, next_morning_df], ignore_index=True)

    # Only charge when price is not above the NEXT day's average
    charge_hours_df = charge_window_df[
        charge_window_df['Price'] <= next_day_avg_price
    ].copy()

    if charge_hours_df.empty:
        avg_charge_price = np.nan
        charge_hours_count = 0
    else:
        avg_charge_price = charge_hours_df['Price'].mean()
        charge_hours_count = len(charge_hours_df)

    daily_rows.append({
        'date': current_date,
        'day_type': 'workday' if pd.Timestamp(current_date).weekday() < 5 else 'weekend',
        'daily_avg_price': daily_avg_price,
        'avg_discharge_price': avg_discharge_price,
        'discharge_hours_count': discharge_hours_count,
        'avg_charge_price': avg_charge_price,
        'charge_hours_count': charge_hours_count
    })

daily_price_df = pd.DataFrame(daily_rows)

# Dynamic spread
daily_price_df['spread'] = (
    daily_price_df['avg_discharge_price'] - daily_price_df['avg_charge_price']
)

daily_price_df.head(10)
```

# Extend Data using Time Series Prediction

The idea is to run the model for a 10 years period from 2021 to 2030. This means we must project existing data into this period.

Existing time series:

- price_df --> 2012/01 to 2026/02
- ev_count_df --> 2008 to 2025
- station_conversion_cost_df --> 2021 and 2030, not the years in between

## Extend Station Conversion Cost

We just need to interpolate between 2021 → 2030 for each cost row and create values for every year.

```python
cost_mid_df = station_conversion_cost_df[['cost', 'mid2021', 'mid2030']].copy()

# reshape to long format
cost_long_df = cost_mid_df.melt(
    id_vars='cost',
    value_vars=['mid2021', 'mid2030'],
    var_name='year',
    value_name='value'
)

cost_long_df['year'] = cost_long_df['year'].str.replace('mid', '').astype(int)

years = pd.DataFrame({'year': list(range(2021, 2031))})
costs = pd.DataFrame({'cost': cost_mid_df['cost'].unique()})

# create full year range for every cost type
full_grid_df = costs.merge(years, how='cross')

cost_interp_df = (
    full_grid_df
    .merge(cost_long_df, on=['cost', 'year'], how='left')
    .sort_values(['cost', 'year'])
)

# interpolate per cost type
cost_interp_df['value'] = (
    cost_interp_df
    .groupby('cost')['value']
    .transform(lambda s: s.interpolate(method='linear'))
)

# pivot back to wide
station_conversion_cost_yearly_df = cost_interp_df.pivot(
    index='cost',
    columns='year',
    values='value'
).reset_index()

display(station_conversion_cost_yearly_df)
save_table(station_conversion_cost_yearly_df, 'forecast_station_conversion_cost_2021_2030', index=False)
```

## Extend EV Count

The exiting data looks exponential. We need an exponential model with an upper cap.

```python
# visualize the growth properly, on both normal scale and log scale.
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Normal scale
axes[0].plot(ev_count_df['Year'], ev_count_df['EV'], marker='o')
axes[0].set_title('EV Count in Hungary (Normal Scale)')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of EV')
axes[0].grid(True)

# Log scale
axes[1].plot(ev_count_df['Year'], ev_count_df['EV'], marker='o')
axes[1].set_yscale('log')
axes[1].set_title('EV Count in Hungary (Log Scale)')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of EV (log)')
axes[1].grid(True)

plt.tight_layout()
save_plot('forecast_ev_count_normal_vs_log')
plt.show()
```

This looks roughly straight, then we can confirm exponential growth is a good model.

```python
# Copy EV count data
ev_projection_df = ev_count_df.copy()

# Logistic growth function
def logistic_model(t, K, A, b):
    return K / (1 + A * np.exp(-b * t))

# Keep only positive EV counts for fitting
ev_fit_df = ev_projection_df[ev_projection_df['EV'] > 0].copy()

# Time index starting from first available year
base_year = ev_fit_df['Year'].min()
ev_fit_df['t'] = ev_fit_df['Year'] - base_year

# Prepare data for fitting
t_data = ev_fit_df['t'].values
y_data = ev_fit_df['EV'].values

# Initial parameter guesses
# K = upper cap, A and b are shape parameters
K0 = 1_000_000
A0 = 1000
b0 = 0.3

# Fit logistic model
params, _ = curve_fit(
    logistic_model,
    t_data,
    y_data,
    p0=[K0, A0, b0],
    maxfev=10000
)

K, A, b = params

print(f"Estimated K (upper cap): {K:,.0f}")
print(f"Estimated A: {A:.4f}")
print(f"Estimated b: {b:.4f}")

# Create projection until 2030
ev_projection_yearly_df = pd.DataFrame({
    'Year': list(range(ev_projection_df['Year'].min(), 2031))
})

ev_projection_yearly_df['t'] = ev_projection_yearly_df['Year'] - base_year
ev_projection_yearly_df['EV_Projected'] = logistic_model(
    ev_projection_yearly_df['t'],
    K, A, b
)

# Round projected EV count
ev_projection_yearly_df['EV_Projected'] = ev_projection_yearly_df['EV_Projected'].round(0).astype(int)

# Merge actual values for comparison
ev_projection_yearly_df = ev_projection_yearly_df.merge(
    ev_count_df[['Year', 'EV']],
    on='Year',
    how='left'
)

# Plot
plt.figure(figsize=(10, 6))

plt.plot(
    ev_projection_yearly_df['Year'],
    ev_projection_yearly_df['EV_Projected'],
    linestyle='--',
    marker='o',
    label='Logistic Projection'
)

plt.plot(
    ev_count_df['Year'],
    ev_count_df['EV'],
    linestyle='-',
    marker='o',
    label='Actual EV Count'
)

plt.title('EV Count Projection in Hungary (Logistic Growth with Upper Cap)')
plt.xlabel('Year')
plt.ylabel('Number of EV')
plt.grid(True)
plt.legend()
plt.tight_layout()
save_plot('forecast_ev_count_logistic_projection')
plt.show()

display(ev_projection_yearly_df.tail(10))
save_table(ev_projection_yearly_df, 'forecast_ev_count_logistic_projection', index=False)
```

```python
# Create a df that use actual data for first 5 years, then use projected data for the second 5 years

ev_count_2021_2030_df = ev_projection_yearly_df[
    (ev_projection_yearly_df['Year'] >= 2021) &
    (ev_projection_yearly_df['Year'] <= 2030)
].copy()

ev_count_2021_2030_df['EV'] = ev_count_2021_2030_df['EV'].fillna(
    ev_count_2021_2030_df['EV_Projected']
)

ev_count_2021_2030_df['EV'] = ev_count_2021_2030_df['EV'].astype(int)

ev_count_2021_2030_df = ev_count_2021_2030_df[['Year', 'EV']]

display(ev_count_2021_2030_df)
save_table(ev_count_2021_2030_df, 'forecast_ev_count_2021_2030_final', index=False)
```

## Extend Electricity Price

For this, we need to use a proper time series prediction with seasonality considerations.

Let's use PyTimeTK: https://business-science.github.io/pytimetk/

```python
# create time series
price_ts_df = price_df.copy()
price_ts_df = price_ts_df.sort_values('Datetime')
price_ts_df = price_ts_df[['Datetime', 'Price']].copy()

# we have some duplicates in our dataset
duplicate_count = price_ts_df['Datetime'].duplicated().sum()
print('Duplicate Datetime rows:', duplicate_count)

price_ts_df[price_ts_df['Datetime'].duplicated(keep=False)].head(10)
```

```python
# aggregate duplicates into one value (average) per timestamp before asfreq
price_ts_df = (
    price_ts_df
    .groupby('Datetime', as_index=False)['Price']
    .mean()
    .sort_values('Datetime')
)

price_ts_df = price_ts_df.set_index('Datetime')
price_ts_df = price_ts_df.asfreq('h')

price_ts_df.head()
```

```python
# Check null values
print('Missing values before fill:', price_ts_df['Price'].isna().sum())

# Fill missing values using time interpolation
price_ts_df['Price'] = price_ts_df['Price'].interpolate(method='time')

# If any nulls remain at the edges, fill them forward/backward
price_ts_df['Price'] = price_ts_df['Price'].bfill().ffill()

# Check again
print('Missing values after fill:', price_ts_df['Price'].isna().sum())

price_ts_df.info()
price_ts_df.head()
```

```python
# We want forecast to last known point: 2030-12-31
last_date = price_ts_df.index.max()
forecast_end = pd.Timestamp('2030-12-31 23:00:00')
n_hours = int((forecast_end - last_date).total_seconds() / 3600)


# Forecast with PyTimeTK: We use seasonality aware model.
price_model_df = price_ts_df.copy().reset_index()

# Add time features
price_model_df = tk.augment_timeseries_signature(
    price_model_df,
    date_column='Datetime'
)

price_model_df.head()
```

```python
# Feature selection
features = [
    'Datetime_hour',
    'Datetime_wday',
    'Datetime_month',
    'Datetime_year',
    'Datetime_yweek',
    'Datetime_yday'
]

target = 'Price'
```

```python
# Train with Random Forest
model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

model.fit(
    price_model_df[features],
    price_model_df[target]
)
```

```python
# Create future dataframe
future_index = pd.date_range(
    start=last_date + pd.Timedelta(hours=1),
    end='2030-12-31 23:00:00',
    freq='h'
)

future_df = pd.DataFrame({
    'Datetime': future_index
})

future_df = tk.augment_timeseries_signature(
    future_df,
    date_column='Datetime'
)

# Predict
future_df['Price_Predicted'] = model.predict(
    future_df[features]
)
```

```python
# Combine actual and forecast
actual_df = price_ts_df.reset_index()

actual_df['Price_Predicted'] = actual_df['Price']

price_forecast_df = pd.concat(
    [
        actual_df[['Datetime', 'Price_Predicted']],
        future_df[['Datetime', 'Price_Predicted']]
    ],
    ignore_index=True
)

display(price_forecast_df)
save_table(price_forecast_df, 'forecast_electricity_price_2015_2030', index=False)
```

```python
# Plot forecast

plt.figure(figsize=(12,5))

plt.plot(
    price_forecast_df['Datetime'],
    price_forecast_df['Price_Predicted'],
    linewidth=1
)

plt.title('Electricity price forecast to 2030')
plt.xlabel('Date')
plt.ylabel('Price EUR/kWh')
plt.grid(True)

save_plot('forecast_electricity_price_to_2030')
plt.show()
```

As we can see, this method keeps:
- daily cycle
- weekly cycle
- yearly cycle
- trend

But does not explode like exponential. This is a conservative chioce for feasibility simulation.
